```python
import re

# modulo para eliminar los simbolos que no sean alfanumericos
def removerPontuacao(texto):
    return re.sub(r'[^A-Za-z0-9 ]', '', texto).strip().lower()
```

```python
import os.path # libreria para ubicar un archivo

arquivo = os.path.join('.', 'pg100.txt') # leemos el archivo pg100.txt que se encuentra en la misma carpeta

# paralelizamos el archivo y eliminamos simbolos que no sean alfanumericos
shakesRDD = (sc.textFile(arquivo, 8).map(removerPontuacao))

# eliminamos lineas en blanco y convertimos cada linea en una lista de palabras
shakesPalavrasRDD = shakesRDD.flatMap(lambda x: x.split())

# filtramos las palabras que tengan una longitud mayor a 3 letras o numeros
shakesPalavrasRDD = shakesPalavrasRDD.filter(lambda x : len(x) > 3)

# eliminamos repetidos 
shakesPalavrasRDD = shakesPalavrasRDD.map(lambda x : (x,1)) # tuplas
shakesPalavrasRDD = shakesPalavrasRDD.groupByKey() # reducimos por clave
shakesPalavrasRDD = shakesPalavrasRDD.map(lambda x: x[0]) # tomamos solo las claves

shakesPalavrasRDD.collect()
```
